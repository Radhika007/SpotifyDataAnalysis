```python
df['duration'] = df['duration_ms'].apply(lambda x: round(x/1000))
df.drop('duration_ms',inplace=True,axis=1)
```

```python
df.head()
```

```python
corr_df = df.drop(['key','mode'],axis=1).corr(method='pearson')
plt.figure(figsize=[14,6])
heatmap = sns.heatmap(corr_df,annot=True,cmap='inferno')
heatmap.set_title('Correlatino Heatmap')
```

```python
# Sampling only 4% of the dataset
sample_df = df.sample(int(0.004*len(df)))
```

```python
# Regression plot
plt.figure(figsize=[10,6])
sns.regplot(data=sample_df,y='loudness',x='energy',color='c')
```

```python
# Correlation between popularity and acoustics
plt.figure(figsize=[10,6])
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='g')
```

```python
#Duration of a song
plt.figure(figsize=[15,6])
plt.title('Duration of song in different genres')
sns.barplot(x='duration',y='genre',data=df)
```

```python
#Top 5 Popular genres
plt.figure(figsize=[10,6])
most_popular_genres = df.sort_values('popularity',ascending=False).head(10)
sns.barplot(x='popularity',y='genre',data=most_popular_genres)
```
